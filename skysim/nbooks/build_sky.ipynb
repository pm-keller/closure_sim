## Build Sky

```python
gleam_sky = gleam_im * beam.beam_image
fg_sky = (fg.eor + gleam_im) * beam.beam_image 
bg_sky = (bg.eor + gleam_im) * beam.beam_image 
```

```python
l_array, m_array, freq = fg.get_sky_coodinates()
coord = np.meshgrid(l_array, m_array)
```

## Compute Visibilities

```python
ant_pos = np.array([[0, 0], [14.6, 0], [7.3, np.sqrt(14.6**2-7.3**2)]])
FGVis = vs.VisData(ant_pos=ant_pos, freq_array=freq_array)
BGVis = vs.VisData(ant_pos=ant_pos, freq_array=freq_array)
Gleam = vs.VisData(ant_pos=ant_pos, freq_array=freq_array)
FGVis.vis_from_image(fg_sky, coord, freq_array)
BGVis.vis_from_image(bg_sky, coord, freq_array)
Gleam.vis_from_image(gleam_im * beam.beam_image, coord, freq_array)
```

## Compute Closure Phases and Plot Fluctuations

```python
fg_eicp = FGVis.get_eicp([0, 1, 2], True)
bg_eicp = BGVis.get_eicp([0, 1, 2], True)
eicp2 = Gleam.get_eicp([0, 1, 2], True)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(6, 4), sharex=True)

ax[0].plot(freq_array.to(u.MHz), np.angle(fg_eicp), color="k")
ax[1].plot(freq_array.to(u.MHz), np.angle(fg_eicp*eicp2.conjugate()), color="blue", label="Faint Galaxies")
ax[1].plot(freq_array.to(u.MHz), np.angle(bg_eicp*eicp2.conjugate()), color="red", label="Bright Galaxies")
ax[1].set_xlabel("Frequency (MHz)")
ax[0].set_ylabel(r"$\phi_\bigtriangledown$ (rad)")
ax[1].set_ylabel(r"$\delta\phi_\bigtriangledown$ (rad)")
ax[0].set_xlim([freq_array[0].to(u.MHz).value, freq_array[-1].to(u.MHz).value])
ax[1].set_ylim([-6e-4, 6e-4])
ax[1].legend(frameon=False)
plt.tight_layout()
plt.savefig("/home/pascal/PhD Project/Code/ClosureSim/plots/closure_spec_delta.png", dpi=200)
```

## Compute and Plot Power Spectra

```python
bh = get_window("blackmanharris", len(freq_array))
bh = bh / np.sqrt(np.mean(bh**2))
bh2 = np.convolve(bh, bh)[::2]
bh2 = bh2 / np.sqrt(np.mean(bh2**2))
```

```python

fs = 1 / np.mean(freq_array[1:]-freq_array[:-1]).to(u.Hz)
DSpec1 = ds.DSpec(eicp2*u.Jy, freq_array=freq_array, window=bh2)
FGSpec = ds.DSpec(fg_eicp*u.Jy, freq_array=freq_array, window=bh2)
BGSpec = ds.DSpec(bg_eicp*u.Jy, freq_array=freq_array, window=bh2)

ps1 = DSpec1.get_xps().apply_cosmo_units(150 * u.m**2)
fg_ps = FGSpec.get_xps().apply_cosmo_units(150 * u.m**2)
bg_ps = BGSpec.get_xps().apply_cosmo_units(150 * u.m**2)

fig, ax = plt.subplots(figsize=(8, 6))
ax = ps1.plot_power(ax=ax, color="k", linestyle="--", linewidth=1.0, label="Foregrounds")
ax = fg_ps.plot_power(ax=ax, linestyle="-", color="b", linewidth=2.0, label=r"Foregrounds + $\mathrm{H\,I}$")
ax = bg_ps.plot_power(ax=ax, linestyle="-", color="r", linewidth=2.0, label=r"Foregrounds + $\mathrm{H\,I}$ (Bright Glx.)")
ax.hlines(74794, -5, -0.18, linestyle="dotted", label="Noise Level IDR3.2 Field C, Band 2")
ax.hlines(74794,  0.18, 5, linestyle="dotted")
ax.set_ylim([1e-5, 1e15])
ax.legend(loc=1)
plt.tight_layout()
plt.savefig("/home/pascal/PhD Project/Code/ClosureSim/plots/closure_ps_sim.png", dpi=200)
```
